### Importar el archivo con datos meteorológicos
Se importa el archivo de datos meteorológicos con periodicidad horaria, generados en R Studio utilizando el parquete METEOSTAT

```python
## Se importa las librerias necesarias
from pandas import read_csv
import pandas as pd
from statistics import mean

## Se importa el dataset
dataset = read_csv('datasetEC.csv', delimiter=';', encoding='cp1252', decimal=',' )
dataset['ws'] = dataset['ws'].fillna(0)
dataset['air_temp'] = dataset['air_temp'].fillna(0)
dataset['visibility'] = dataset['visibility'].fillna(0)
dataset['RH'] = dataset['RH'].fillna(0)
dataset['State'] = dataset['State'].fillna(0)
dataset.tail()
```

```python
## se toma información filtrando un rango de fechas
dataset["date"]=pd.to_datetime(dataset["date"])
dataset["date2"]=pd.to_datetime(dataset['date']).dt.normalize()    
dat=dataset.loc[dataset["date2"].between('2021-01-16', '2021-04-15')]
```

```python
## obtener los datos diarios de WS velocidad del viento en m/s
a=pd.DataFrame(dat)
b=a[a.ws!=0]
df2 = b.groupby(['date2', 'station','State']).ws.mean()

# crear una lista con los valores agrupados y la media de WS velocidad del viento
lista = [list(key) + [str(value)] for key,  value in  df2.to_dict().items()]

# crear un dataframe y renombrar los campos
df2 = pd.DataFrame(lista)
df2.rename(columns={0: 'date',1:'station', 2:'state', 3:'ws'}, inplace=True)
df2.tail()
```

```python
## obtener los datos diarios de air_temp temperatura del aire °C
c=a[a.air_temp!=0]
df3 = c.groupby(['date2', 'station','State']).air_temp.mean()

# crear una lista con los valores agrupados y la media de air_temp temperatura del aire
lista2 = [list(key) + [str(value)] for key,  value in  df3.to_dict().items()]

# crear un dataframe y renombrar los campos
df3 = pd.DataFrame(lista2)
df3.rename(columns={0: 'date',1:'station_2', 2:'state', 3:'air_temp'}, inplace=True)
df3.tail()
```

```python
## obtener los datos diarios de visibility en metros
d=a[a.visibility!=0]
df4 = d.groupby(['date2', 'station','State']).visibility.mean()

# crear una lista con los valores agrupados y la media de visibility visibilidad
lista3 = [list(key) + [str(value)] for key,  value in  df4.to_dict().items()]

# crear un dataframe y renombrar los campos
df4 = pd.DataFrame(lista3)
df4.rename(columns={0: 'date',1:'station_3', 2:'state', 3:'visibility'}, inplace=True)
df4.tail()
```

```python
## obtener los datos diarios de RH Humedad Relativa en %
e=a[a.RH!=0]
df5 = e.groupby(['date2', 'station','State']).RH.mean()

# crear una lista con los valores agrupados y la media de air_temp temperatura del aire
lista4 = [list(key) + [str(value)] for key,  value in  df5.to_dict().items()]
df5 = pd.DataFrame(lista4)

# Se renombra los atributos
df5.rename(columns={0: 'date',1:'station_4', 2:'state', 3:'RH'}, inplace=True)
df5.tail() 
```

```python
# se unen los datsets
df2_3= pd.merge(df2, df3, on=['date','state'], how='right')
df23_4= pd.merge(df2_3, df4, on=['date','state'], how='left')
data= pd.merge(df23_4, df5, on=['date','state'], how='left')

# Se eliminan los atributos del dataset
data.drop(['station_2','station_3','station_4'], axis='columns', inplace=True)

# Se rellena con 0 los campos con valores NA
data['ws'] = data['ws'].fillna(0)
data['visibility'] = data['visibility'].fillna(0)
data['RH'] = data['RH'].fillna(0)

data
```

```python
# conversión de tipo de datos
data['ws']=pd.to_numeric(data['ws'])
data['air_temp']=pd.to_numeric(data['air_temp'])
data['visibility']=pd.to_numeric(data['visibility'])
data['RH']=pd.to_numeric(data['RH'])
data['date']=pd.to_datetime(data['date']).dt.normalize()

# conversión de valor ws (m/s) --> mph
valor_conv_ws=2.24 
data['ws']=data['ws']*valor_conv_ws

# conversión de valor visibility (m) --> millas
valor_conv_vis=0.00062137
data['visibility']=data['visibility']*valor_conv_vis
```

### Importar el archivo con datos epidemiológicos
Se importa el archivo de datos epidemiológicos con periodicidad diaria, obtenidos desde la plataforma github.

```python
## Se importa el dataset
dataset2 = read_csv('DATOS_ECUADOR.csv', delimiter=';', encoding='cp1252')

# Se crea una lista con los atributos que no son usados para obtener las variables metereológicas
atributos_eliminar=['inec_provincia_id','poblacion', 'total','lat','lng']
# Se eliminan los atributos del dataset
for atributo in atributos_eliminar:
    dataset2.drop(atributo, axis='columns', inplace=True)
    
# se renombra los campos
dataset2.rename(columns={'provincia': 'state','nuevas':'new_cases', 'created_at':'date'}, inplace=True)

# se cambia el tipo de datos de la fecha
dataset2['date']=pd.to_datetime(dataset2['date']).dt.normalize()

# se filtra el rango de fechas para el análisis
dataset2=dataset2.loc[dataset2["date"].between('2021-01-16', '2021-04-15')]
dataset2.tail()
```

### Creación de dataset con latencias 0, 3, 7, 14 días 
Estos datasets incluyen datos meteorológicos y datos epidemiológicos con periodicidad diaria.

### 1. Latencia 0 días

```python
## UNIR LOS DATASETS
dataset_sin_latencia= pd.merge(data, dataset2, on=['date','state'], how='inner')

dataset_sin_latencia.tail()
```

### 2. Latencia 3 días

```python
## dataset con latencia 3 días
from datetime import date, timedelta
data1=pd.DataFrame(data)
data2=pd.DataFrame(data)

# se declara el número de días de retardo
td = timedelta(3)
data1['date1']=data1['date']-td

# Se crea una lista con los atributos que no son usados para obtener las variables metereológicas
col_eliminar=['station','ws','air_temp', 'visibility', 'RH']
# Se eliminan los atributos del dataset
for atributo in col_eliminar:
    data1.drop(atributo, axis='columns', inplace=True)

## se calcula obtiene las variables meteorológicas para un retardo de 3 días
data_3dias_latencia=pd.merge(data1, data2, left_on=['date1','state'], right_on=['date','state'], how='inner')

## Eliminar la columna date usadas para obtener retardo 3 días
data_3dias_latencia.drop(['date1_x', 'date_y','date1_y'], axis='columns', inplace=True)

# se renombra el campo date 
data_3dias_latencia.rename(columns={'date_x': 'date'}, inplace=True)

# se filtra el rango de fechas para el análisis
data_3dias_latencia=data_3dias_latencia.loc[data_3dias_latencia["date"].between('2021-01-16', '2021-04-15')]

data_3dias_latencia=pd.merge(data_3dias_latencia, dataset2, left_on=['date','state'], right_on=['date','state'], how='inner')
data_3dias_latencia.tail()
```

### 3. Latencia 7 días

```python
## dataset con latencia 7 días
data3=pd.DataFrame(data)

# número de dias de retardo
td = timedelta(7)

# se obtiene las fechas con retardo
data3['date1']=data3['date']-td

# Se eliminan los atributos del dataset
for atributo in col_eliminar:
    data3.drop(atributo, axis='columns', inplace=True)

## se calcula obtiene las variables meteorológicas para un retardo de 7 días
data_7dias_latencia=pd.merge(data3, data2, left_on=['date1','state'], right_on=['date','state'], how='inner')

## Eliminar la columna date usadas para obtener retardo 3 días
data_7dias_latencia.drop(['date1_x', 'date_y','date1_y'], axis='columns', inplace=True)

# se renombra el campo date 
data_7dias_latencia.rename(columns={'date_x': 'date'}, inplace=True)

# se filtra el rango de fechas para el análisis
data_7dias_latencia=data_7dias_latencia.loc[data_7dias_latencia["date"].between('2021-01-16', '2021-04-15')]

data_7dias_latencia=pd.merge(data_7dias_latencia, dataset2, left_on=['date','state'], right_on=['date','state'], how='inner')

data_7dias_latencia.tail()
```

### 4. Latencia 14 días

```python
## dataset con latencia 14 días
data4=pd.DataFrame(data)

# se define número de dias de retardo
td = timedelta(14)

# se obtiene las fechas con retardo
data4['date1']=data4['date']-td

# Se eliminan los atributos del dataset
for atributo in col_eliminar:
    data4.drop(atributo, axis='columns', inplace=True)

## se calcula obtiene las variables meteorológicas para un retardo de 3 días
data_14dias_latencia=pd.merge(data4, data2, left_on=['date1','state'], right_on=['date','state'], how='inner')

## Eliminar la columna date usadas para obtener retardo 3 días
data_14dias_latencia.drop(['date1_x', 'date_y','date1_y'], axis='columns', inplace=True)

# se renombra el campo date 
data_14dias_latencia.rename(columns={'date_x': 'date'}, inplace=True)

# se filtra el rango de fechas para el análisis
data_14dias_latencia=data_14dias_latencia.loc[data_14dias_latencia["date"].between('2021-01-16', '2021-04-15')]

## se obtiene el dataset completo para latencia de 14 dias
data_14dias_latencia=pd.merge(data_14dias_latencia, dataset2, left_on=['date','state'], right_on=['date','state'], how='inner')

data_14dias_latencia.tail()
```

#### Se exporta los dataset generados a archivos .csv

```python
#### Se eliminan los registros con valor 0 en el campo visibilidad 
data_14dias_latencia=data_14dias_latencia[data_14dias_latencia.visibility!=0]
data_7dias_latencia=data_7dias_latencia[data_7dias_latencia.visibility!=0]
data_3dias_latencia=data_3dias_latencia[data_3dias_latencia.visibility!=0]
dataset_sin_latencia=dataset_sin_latencia[dataset_sin_latencia.visibility!=0]

## Se crea archivos csv con datos de las 3 latencias, respectivamente
data_14dias_latencia.to_csv('dataset_14dias_retardo.csv', index=False,encoding="cp1252")
data_7dias_latencia.to_csv('data_7dias_latencia.csv', index=False,encoding="cp1252")
data_3dias_latencia.to_csv('data_3dias_latencia.csv', index=False,encoding="cp1252")
dataset_sin_latencia.to_csv('dataset_sin_latencia.csv', index=False,encoding="cp1252")
```
